```python
from shapely.wkt import loads
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import contextily

# Read the CSV file into a DataFrame
df1 = pd.read_csv("schiphol_noise_night.csv")

# Initialize an empty list to store geometries
geometries = []

# Iterate through each row in the DataFrame
for idx, row in df1.iterrows():
    multipolygon_wkt = row["geom"]
    multipolygon = loads(multipolygon_wkt)
    geometries.append(multipolygon)

# Create a GeoDataFrame from the list of geometries
gdf = gpd.GeoDataFrame(geometry=geometries, crs='EPSG:28992')

# Add 'db' and other columns to the GeoDataFrame
gdf['db'] = df1['fromto']
gdf['area'] = gdf.area / 1000000
gdf['centroid'] = gdf.centroid
gdf['boundary'] = gdf.boundary

# Plot with custom colors using a colormap and smoother edges
ax = gdf.plot(column='db', figsize=(10, 10), aspect=1, edgecolor='grey', linewidth=0.5, legend=True, cmap='autumn_r')

# Add a basemap with appropriate zoom level and attribution
contextily.add_basemap(ax, crs=gdf.crs.to_string(), zoom=12, attribution="© OpenStreetMap contributors")

# Set the title and turn off the axis
plt.title('Noise pollution around Schiphol')
plt.axis('off')
plt.show()
```
